```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import math as m
from tqdm import tqdm
```

```python
mu = 1.327124400188e20
r_1 = 149.6e9
r_f = 5.79e10
```

```python
table_vh = pd.DataFrame(index=pd.RangeIndex(100),columns=pd.RangeIndex(1))
table_vh
```

```python
for i in tqdm(range(100)):
    r_2 = 0.01*r_1 + i*r_1/100
    dv_1 = abs(np.sqrt(mu/r_1)*(((2*r_2)/(r_1+r_2))**0.5-1))
    dv_2 = abs(np.sqrt(mu/r_2)*(1-((2*r_1)/(r_1+r_2))**0.5))
    v_sum = (dv_1 + dv_2)/1000
    table_vh.loc[i]=v_sum
```

```python
table_vh
```

```python
table_v = pd.DataFrame(index=pd.RangeIndex(100),columns=pd.RangeIndex(3000))
table_v
```

```python
for j in tqdm(range(3000)):
    for i in range(100):
        r_2 = 0.01*r_1 + i*r_1/100
        r_b = 0.01*r_1 + j*r_1/100
        a_1 = (r_1+r_b)/2
        a_2 = (r_2+r_b)/2
        dd_v1 = abs(np.sqrt((2*mu/r_1)-mu/a_1)-np.sqrt(mu/r_1))
        dd_v2 = abs(np.sqrt((2*mu/r_b)-mu/a_2)-np.sqrt((2*mu/r_b)-mu/a_1))
        dd_v3 = abs(np.sqrt((2*mu/r_2)-mu/a_2)-np.sqrt(mu/r_2))
        summm = (dd_v1+dd_v2+dd_v3)/1000
        table_v.loc[i,j]=summm
```

```python
table_v
```

```python
table_r = pd.DataFrame(index=pd.RangeIndex(100),columns=pd.RangeIndex(3000))
table_r
```

```python
for i in tqdm(range(100)):
    for j in range(3000):
        table_r.loc[i,j] = table_v.loc[i,j] - table_vh.loc[i,0]
```

```python
table_r
```

```python
%matplotlib notebook
import numpy as np
from scipy.interpolate import RegularGridInterpolator

z = table_r.to_numpy()

x1 = np.linspace(0, 99, 100)
x2 = np.linspace(0, 2999, 3000)

f = RegularGridInterpolator((x1, x2), z, bounds_error=False, fill_value=None)
def convenient_f(x1, x2):
    x = np.vstack((x1.ravel(), x2.ravel())).T
    z = f(x)
    return z.reshape(x1.shape)

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import pyplot as plt

X1, X2 = np.meshgrid(x1, x2, indexing='ij')
x1_surface, x2_surface = np.meshgrid(np.linspace(0, 99, 100), np.linspace(0, 3000, 3001), indexing='ij')

fig = plt.figure(figsize=(10,8))

ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x1_surface, x2_surface, convenient_f(x1_surface, x2_surface), alpha=0.5)
ax.view_init(10, 80)

for angle in range(0, 360):
    ax.view_init(30, angle)

plt.show()
```

```python
# Будем искать в каждой строчке, с какого элемента начинаются отрицательные, то есть с какого момента
# биэллиптические переходы будут устойчиво эффективнее. 

table_negative = pd.DataFrame(index=pd.RangeIndex(100),columns=pd.RangeIndex(1))


for i in tqdm(range(100)):
    first_negative = None
    for j in range(2999):
        
        if (table_r.loc[i,j] >= 0) and (table_r.loc[i,j+1] < 0):
            first_negative = j+1
        if (table_r.loc[i,j] < 0) and (table_r.loc[i,j+1] >= 0):
            first_negative = None
    table_negative.loc[i]=first_negative
```
